```python
from scipy.stats import norm 
from numpy import arange
import matplotlib.pyplot as plt
plt.style.use('bmh')
%matplotlib inline
```

```python
plt.figure(figsize=(12, 12))
for s in arange(0.55, 1.15, 0.15):
    Pds = [1 - norm.cdf((l-1)/s) for l in arange(0, 1.01, 0.04)]
    Pfas = [1 - norm.cdf(l/s) for l in arange(0, 1.01, 0.04)]
    ax = plt.plot(Pfas, Pds)
    x, y = ax[0].get_xydata()[13]
    plt.annotate(f'$\sigma={s:.4}$', (x + 0.03, y), 
                 fontsize=30, color='black', rotation=40)
plt.xlabel('$P_{fa}$', fontsize=35, color='k')
plt.ylabel('$P_{d}$', fontsize=35, color='k', rotation=0)
plt.title('ROC-кривая в зависимости от $\sigma$', fontsize=20)
```

```python
for s in arange(0.55, 1.15, 0.15):
    plt.figure(figsize=(10, 6))
    d1, d2 = [], []
    xs = arange(-5, 5, 0.01)
    for x in xs:
        d1.append(norm.pdf(x/s))
        d2.append(norm.pdf((x-1)/s))
    ax = plt.plot(xs, d1, color='green', linewidth=5)
    ax2 = plt.plot(xs, d2, color='red', linewidth=5)
    coord = ax[0].get_xydata()[25]
    plt.annotate(f'$\sigma={s:.2f}$', (coord[0] - 0.01, coord[1]+0.3), 
                 fontsize=30)
    plt.xlabel('$x$', fontsize=35, color='k')
    plt.ylabel('$f$', fontsize=35, color='k', rotation=0, labelpad=10)
    plt.title('Условные плотности распределений', fontsize=20)
    plt.vlines(0, 0, 0.4, linestyles='--')
    plt.vlines(1, 0, 0.4, linestyles='--')
    plt.show()
```

```python

from numpy import arange
from scipy.stats import norm 
import matplotlib.pyplot as plt
plt.style.use('grayscale')

%matplotlib inline
```

```python
arange(0.5, 1.1, 0.3)
```

```python
plt.figure(figsize=(12, 12))
for s in arange(0.5, 1.1, 0.3):
    Pds, Pfas = [], []
    for l in arange(0, 1.01, 0.04):
        Pfas.append(1 - norm.cdf(l/s))
        Pds.append(1 - norm.cdf((l-1)/s))
    ax = plt.plot(Pfas, Pds, linewidth=5)
    coord = ax[0].get_xydata()[12]
    plt.annotate(f'$(\sigma={s:.4})$', (coord[0] + 0.04, coord[1]+ 0.01),
                 color='gray', fontsize=40)
#     plt.tick_params(axis='x', colors='black')
#     plt.tick_params(axis='y', colors='red')
```

```python
fig, axs = plt.subplots(1, 4, figsize=(18, 5))

for i, s in enumerate(arange(0.5, 1.2, 0.2)):
    plt.figure(figsize=(12, 4))
    xs = arange(-5, 5, 0.01)
    d1, d2 = [norm.pdf(x/s) for x in xs], [norm.pdf((x-1)/s) for x in xs]
    axs[i%4].plot(xs, d1, color='b', linewidth=5)
    axs[i%4].plot(xs, d2, color='r', linewidth=5)
    axs[i%4].set_title(f'$\sigma={s:.1f}$', fontsize=20);
```
